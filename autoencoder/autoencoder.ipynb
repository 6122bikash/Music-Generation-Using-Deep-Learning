```python
import midi
```

```python
directories=['session']
midi.get_instruments_list(directories)
```

```python
sample=midi.midi_datasets_to_samples(directories,48,save_at='sample.npy')
```

```python
X=sample[0:4*(sample.shape[0]//4),:,:]
```

```python
X=X.reshape(-1,4,96,96)
X_shape=X.shape
```

```python
from keras.layers import Input,Dense
from keras.models import Model
from keras import regularizers
```

```python
encoding_dim=96
input_notes=Input(shape=(4*96*96,))
encoded=Dense(encoding_dim,activation='relu',activity_regularizer=regularizers.l1(0.000001))(input_notes)
decoded=Dense(4*96*96,activation="sigmoid")(encoded)
autoencoder=Model(input_notes,decoded)
```

```python
#seperate encoded model
#mapping input to encoded representation
encoder=Model(input_notes,encoded)
#seperate decoder model
encoded_input=Input(shape=(encoding_dim,))
#last layer of auto encoded model
decoder_layer=autoencoder.layers[-1]
#decoder model
decoder=Model(encoded_input,decoder_layer(encoded_input))
```

```python
autoencoder.compile(optimizer='adadelta',loss='binary_crossentropy')
```

```python
import numpy as np
x_train=X.astype('float32')/210.
x_train=x_train.reshape(len(x_train),np.prod(x_train.shape[1:]))
print(x_train.shape)
```

```python

history=autoencoder.fit(x_train,x_train,epochs=10,batch_size=256,shuffle=True,validation_split=0.1)
```

```python
import matplotlib.pyplot as plt
val_loss=history.history['val_loss']
loss=history.history['loss']
plt.plot(val_loss,label='val_loss')
plt.plot(loss,label='loss')
plt.legend()
```
